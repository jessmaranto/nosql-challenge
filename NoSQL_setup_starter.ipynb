5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
documents = establishments.find({})
for document in documents:
    if 'geocode' in document:
        longitude_str = document['geocode'].get('longitude')
        latitude_str = document['geocode'].get('latitude')
        
        # Convert to Float if they are not None
        if longitude_str is not None and latitude_str is not None:
            try:
                longitude_float = float(longitude_str)
                latitude_float = float(latitude_str)
                
                # Update the document with the new Float values
                establishments.update_one(
                    {'_id': document['_id']},  # Match the document by its unique ID
                    {'$set': {
                        'geocode.longitude': longitude_float,
                        'geocode.latitude': latitude_float
                    }}
                )
            except ValueError:
                print(f"Could not convert longitude or latitude for document ID: {document['_id']}")

print("Longitude and latitude have been updated to Float.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
for document in documents:
    # Check if 'RatingValue' exists in the document
    if 'RatingValue' in document:
        rating_value_str = document['RatingValue']
        
        # Check if the RatingValue is a valid integer string
        if isinstance(rating_value_str, str) and rating_value_str.isdigit():
            rating_value_int = int(rating_value_str)
            
            # Update the document with the new Integer value
            establishments.update_one(
                {'_id': document['_id']},  # Match the document by its unique ID
                {'$set': {'RatingValue': rating_value_int}}
            )
        else:
            print(f"Skipping document ID {document['_id']} with invalid RatingValue: {rating_value_str}")
    else:
        print(f"Document ID {document['_id']} does not have a RatingValue field.")
```

```python
# Check that the coordinates and rating value are now numbers
[doc for doc in establishments.find({}, ['geocode.longitude', 'geocode.latitude', 'RatingValue'])]
```
